```python
from google.cloud import speech
key = '/content/drive/My Drive/Curioson-personal/google-key.json'
#Google cloud service url to bucket
wav9gcs = 'gs://benchmark-audio/9.wav'
wav17gcs = 'gs://benchmark-audio/17.wav'

client = speech.SpeechClient.from_service_account_file(key)
```

```python
##17.wav

audio = speech.RecognitionAudio(uri=wav17gcs)

config = speech.RecognitionConfig(
  enable_automatic_punctuation=True,
  language_code='en-US'
)

operation = client.long_running_recognize(config=config, audio=audio)

print("Waiting for operation to complete...")
%%time
response = operation.result(timeout=500)

transcript = []

for result in response.results:
  # The first alternative is the most likely one for this portion.
  transcript.append(result.alternatives[0].transcript)



print(transcript)
```

```python
from deepgram import Deepgram
import asyncio, json, os

dg_key = '6dca77ad87f5407e2b69da04399fd8f39333dc49'
dg = Deepgram(dg_key)

MIMETYPE = 'wav'
DIRECTORY = 'sample_data/benchmark-audio'

options = {
    "punctuate": True,
    "model": 'general',
    "tier": 'enhanced'
}


#This function is what calls on the model to transcribe
def main():
    audio_folder = os.listdir(DIRECTORY)
    for audio_file in audio_folder:
        if audio_file.endswith(MIMETYPE):
          with open(f"{DIRECTORY}/{audio_file}", "rb") as f:
              source = {"buffer": f, "mimetype":'audio/'+MIMETYPE}
              res = dg.transcription.sync_prerecorded(source, options)
              with open(f"./{audio_file[:-4]}.json", "w") as transcript:
                  json.dump(res, transcript)
    return


# The JSON is loaded with information, but if you just want to read the
# transcript, run the code below!
def print_transcript(transcription_file):
  with open(transcription_file, "r") as file:
        data = json.load(file)
        result = data['results']['channels'][0]['alternatives'][0]['transcript']
        result = result.split('.')
        for sentence in result:
          print(sentence + '.')
```

```python
#9 wav
DIRECTORY = 'sample_data/benchmark-audio'
```

```python
%%time
result = main()
print(result)
```

```python
OUTPUT = '9.json'

print_transcript(OUTPUT)
```
